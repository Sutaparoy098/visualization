```python
#1
import seaborn as sns
import pandas as pd
import numpy as np

# Generate synthetic data
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Create a scatter plot
sns.scatterplot(data=df, x='x', y='y')
plt.title('Scatter Plot of Synthetic Data')
plt.show()
```

```python
#2
# Generate random data
data_dist = np.random.randn(1000)

# Create a distribution plot using Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(data_dist, kde=True, color='skyblue')
plt.title('Distribution Plot of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#3
# Create a DataFrame with categories and values
data = {'Category': ['A', 'B', 'C', 'D'], 'Value': [25, 40, 30, 35]}
df = pd.DataFrame(data)

# Create a bar plot
sns.barplot(x='Category', y='Value', data=df)
plt.title('Comparison of Categories')
plt.show()
```

```python
#4
# Create a DataFrame with categories and numerical values
data = {'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'D'],
        'Value': [25, 40, 30, 35, 20, 28, 32]}
df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of Values Across Categories')
plt.show()
```

```python
#5 Generate a synthetic dataset with correlated features
np.random.seed(42)
x1 = np.random.randn(100)
x2 = 0.8 * x1 + np.random.randn(100)
x3 = -0.5 * x1 + np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})

# Create a heatmap
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#1
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Display Bokeh plots in the notebook
output_notebook()

# Create a sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p1 = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y')
p1.line(x, y, line_width=2, color='blue', legend_label='y = sin(x)')

# Show the plot
show(p1)

```

```python
#2
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import pandas as pd

# Generate random data for scatter plot
np.random.seed(42)
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)
sizes = np.random.randint(10, 30, 100)  # Marker sizes
colors = np.random.choice(['red', 'blue', 'green', 'purple'], 100)  # Marker colors

# Create a dataframe and a ColumnDataSource
scatter_data = pd.DataFrame({'x': x_scatter, 'y': y_scatter, 'size': sizes, 'color': colors})
source = ColumnDataSource(scatter_data)

# Create a Bokeh scatter plot
p2 = figure(title="Scatter Plot with Random Values", x_axis_label='X', y_axis_label='Y')
p2.scatter('x', 'y', size='size', color='color', source=source, legend_field='color', alpha=0.6)

# Show the plot
show(p2)
```

```python
#3
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
bar_data = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh figure for the bar chart
p3 = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')
p3.vbar(x='fruits', top='counts', width=0.5, source=bar_data, legend_field='fruits',
        line_color='white', fill_color=factor_cmap('fruits', palette=['red', 'orange', 'yellow', 'green'], factors=fruits))

# Show the plot
show(p3)
```

```python
#4
from bokeh.plotting import figure, show

# Generate random data for the histogram
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure for the histogram
p4 = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')
p4.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.5)

# Show the plot
show(p4)

```

```python
#5
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
import numpy as np

# Generate random data for heatmap
data_heatmap = np.random.rand(10, 10)

# Create meshgrid
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a Bokeh figure for the heatmap
p5 = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', x_range=(0, 1), y_range=(0, 1))

# Create a color mapper
color_mapper = LinearColorMapper(palette='Viridis256', low=np.min(data_heatmap), high=np.max(data_heatmap))

# Add heatmap image
p5.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p5.add_layout(color_bar, 'right')

# Show the plot
show(p5)
```

```python
#1
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter Plot Dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df_3d = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(df_3d['X'], df_3d['Y'], df_3d['Z'], c=df_3d['Z'], cmap='viridis', marker='o')

# Set labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot')

# Add color bar to show scale
fig.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
#2
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Student Grades Dataset
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data)

# Create a violin plot using Seaborn
plt.figure(figsize=(8, 6))
sns.violinplot(x='Grade', y='Score', data=df_grades, palette='pastel')

# Add title and labels
plt.title('Distribution of Scores across Different Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sales Data Dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

# Pivot the data to create a heatmap structure
sales_pivot = df_sales.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, annot=True, fmt=".0f", cmap='YlGnBu', cbar=True)

# Add title and labels
plt.title('Sales Variation across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

```python
#4
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
import numpy as np
import pandas as pd

# Display Bokeh plots inline in the notebook
output_notebook()

# Generate the meshgrid and data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the meshgrid arrays and create a DataFrame
data = {'X': x.flatten(), 'Y': y.flatten(), 'Z': z.flatten()}
df_hist = pd.DataFrame(data)

# Create histogram data for the Z values
hist, edges = np.histogram(df_hist['Z'], bins=30)

# Create a Bokeh figure
p = figure(title="Bokeh Histogram of Z values", x_axis_label='Z', y_axis_label='Frequency')

# Create a quad plot for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.7)

# Show the plot
show(p, notebook_handle=True)
```

```python
#5
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap
import pandas as pd
import numpy as np

# Enable inline Bokeh display in notebooks
output_notebook()

# Random seed for reproducibility
np.random.seed(25)

# Dataset for the Bubble Chart
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df_bubble = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df_bubble)

# Define color mapping based on the population
color_mapper = linear_cmap(field_name='Population', palette='Viridis256', low=df_bubble['Population'].min(), high=df_bubble['Population'].max())

# Create a Bokeh figure
p = figure(x_axis_label='GDP', y_axis_label='Population', title='Bubble Chart: Population vs GDP')

# Create a scatter plot where the size of the circle is proportional to Population
p.scatter(x='GDP', y='Population', size='Population', source=source, fill_alpha=0.6, color=color_mapper, legend_field='Country')

# Import and add HoverTool
hover = HoverTool(tooltips=[("Country", "@Country"), ("Population", "@Population"), ("GDP", "@GDP")])
p.add_tools(hover)

# Show the plot
show(p)
```
